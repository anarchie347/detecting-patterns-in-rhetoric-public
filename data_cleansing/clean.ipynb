```python
import pandas as pd
import os
from pathlib import Path
```

```python
AI_LABEL_NAME = "label0"
NOT_AI_LABEL = 0
IS_AI_LABEL = 1

BULLSHIT_LABEL_NAME = "label1"
NOT_BULLSHIT_LABEL = 0
IS_BULLSHIT_LABEL = 1
```

```python
BASE_DIR = "../"
if not os.path.isdir("data"):
    os.makedirs("data")
```

```python
def process_data_files(input_dir, process_text, ai_label, bullshit_label, output_file):
    contents = []
    for f in Path(input_dir).iterdir():
        try:
            if f.is_file():
                text = process_text(f.read_text())
                contents.append(text)
        except UnicodeDecodeError as e:
            print(str(e.object)[:100]) # ignoring files with unicode errors.
    data = pd.DataFrame({
        "text": contents,
        AI_LABEL_NAME: ai_label,
        BULLSHIT_LABEL_NAME: bullshit_label
    })
    data.to_csv(output_file, index=False)
    return data
```

### Gemini Data

AI Data

```python
import io
import zipfile

def texts_from_zip(zip_path):
    texts = []
    with zipfile.ZipFile(zip_path) as zf:
        entries = [name for name in zf.namelist() if not name.endswith("/")]
        total = len(entries)
        for idx, name in enumerate(entries, start=1):
            with zf.open(name) as fh:
                data = fh.read()

            if name.lower().endswith(".txt"):
                text = data.decode().strip()
                texts.append(text)

            if name.lower().endswith(".csv"):
                try:
                    df = pd.read_csv(io.BytesIO(data), on_bad_lines="skip")
                except:
                    continue

                if "text" in df.columns > 0:
                    value = str(df["text"].iloc[0]).strip()
                elif len(df.columns) == 1:
                    value = str(df.columns[0]).strip()
                else:
                    value = ""

                if value:
                    texts.append(value)

    return texts

bs_zip = f"{BASE_DIR}ai_detector_data/gemini_results_bs.zip"
nbs_zip = f"{BASE_DIR}ai_detector_data/gemini_results_nbs.zip"

bs_texts = texts_from_zip(bs_zip)
nbs_texts = texts_from_zip(nbs_zip)

ai_df = pd.DataFrame({
    "text": bs_texts + nbs_texts,
    AI_LABEL_NAME: IS_AI_LABEL,
    BULLSHIT_LABEL_NAME: [IS_BULLSHIT_LABEL] * len(bs_texts) + [NOT_BULLSHIT_LABEL] * len(nbs_texts),
}).reset_index(drop=True)

ai_df.to_csv("data/ai_gemini.csv", index=False)
ai_df.info()
ai_df.head()
```

### GPT Data

AI Data

```python
import zipfile

gpt_zip = f"{BASE_DIR}ai_detector_data/gpt_articles.zip"
texts = []

def process(text):
    lines = text.splitlines()
    content = []
    for line in lines:
        if line.startswith("###") or line.startswith("**"):
            continue
        content.append(line)
    return "\n".join(content).strip()

with zipfile.ZipFile(gpt_zip) as zf:
    entries = [name for name in zf.namelist() if name.lower().endswith(".txt")]
    total = len(entries)
    for idx, name in enumerate(entries, start=1):
        with zf.open(name) as fh:
            text = fh.read().decode()
            texts.append(process(text))

ai_df = pd.DataFrame({
    "text": texts,
    AI_LABEL_NAME: IS_AI_LABEL,
    BULLSHIT_LABEL_NAME: NOT_BULLSHIT_LABEL,
}).reset_index(drop=True)

ai_df.to_csv("data/ai_gpt.csv", index=False)
ai_df.info()
ai_df.head()
```
